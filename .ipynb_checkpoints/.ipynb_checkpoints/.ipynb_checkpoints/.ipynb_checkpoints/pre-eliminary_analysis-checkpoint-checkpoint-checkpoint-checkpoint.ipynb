```r
#library loading section

library(tidyverse)
library(dplyr)
library(tidymo)
```

```r
#url section

cleveland_url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/cleve.mod"
```

```r
#reading the data

cleveland_data <- read_delim(cleveland_url, delim = " ", col_names = FALSE, skip = 19)

glimpse(cleveland_data)
```

```r
#adding the columns names from the source file

colnames(cleveland_data) <- c("age", "sex", "chest_pain_type", "resting_bp", "cholestrol", "fasting_bs", "resting_ecg", "max_hr", "exercise_angina", "oldpeak", "st_slope", "colored_vessels", "thal", "heart_disease", "sickness_level")
glimpse(cleveland_data)
```

```r
is.na(cleveland_data) # no missing data
```

```r
cleveland_select <- select(cleveland_data, age, chest_pain_type, cholestrol, resting_ecg, oldpeak, heart_disease)
```

```r
cleveland_split <- initial_split(cleveland_select, prop = 0.75, strata = heart_disease)
cleveland_train <- training(cleveland_split)
cleveland_test <- testing(cleveland_split) 
```

```r
# summary of the training data
summary(cleveland_train)
```

```r
#visualizing the relationship between age and cholestrol levels

age_vs_cholestrol_plot <- cleveland_train %>%
                            ggplot(aes(x = age, y = cholestrol)) +
                            geom_point() + 
                            theme_bw() +
                            labs(x = "Age in years", y = "Cholestrol [mm/dl]") +
                            ggtitle("Age vs Cholestrol Levels")

age_vs_cholestrol_plot 
```
